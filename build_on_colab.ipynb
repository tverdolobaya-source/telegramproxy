```python
# Скачать APK
from google.colab import files
import os

apk_files = [f for f in os.listdir('bin') if f.endswith('.apk')]
if apk_files:
    files.download(f'bin/{apk_files[0]}')
    print(f'APK готов: {apk_files[0]}')
else:
    print('APK не найден')
```
